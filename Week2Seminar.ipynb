```python
# Password Strength Checker <---------------------
import re

input_password = input('Enter a password: ')

def password_strength(input_password):
    regex = re.compile('[@_!#$%^&*()<>?/\\\\|}{~:]')
    count = 0
    if len(input_password) > 8:
        count += 1
    else:
        print('The password is not more than 8 characters')
    if any(i.isupper() for i in input_password):
        count += 1
    else:
        print('The password does not contain an uppercase letter')
    if any(i.islower() for i in input_password):
        count += 1
    else:
        print('The password does not contain a lowercase letter')
    if regex.search(input_password):
        count += 1
    else:
        print('The password does not contain a special character')
    if count == 4:
        print('Strong password')
    else:
        print('Weak password')

password_strength(input_password)
```

```python
# Hangman Game <---------------------
import random

word_list = ['python', 'java', 'javascript']

def select_word(word_list):
    return random.choice(word_list)

def display_word(word, guessed_letters):
    return ' '.join([letter if letter in guessed_letters else '_' for letter in word])

def hangman(word_list, max_guesses):
    word = select_word(word_list)
    guessed_letters = []
    incorrect_guesses = 0

    print("Welcome to Hangman!")
    print(display_word(word, guessed_letters))

    while incorrect_guesses < max_guesses:
        guess = input("Guess a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single alphabetic character.")
            continue

        if guess in guessed_letters:
            print("You already guessed that letter.")
        elif guess in word:
            guessed_letters.append(guess)
            print(display_word(word, guessed_letters))
            if all(letter in guessed_letters for letter in word):
                print(f"Congratulations, you guessed the word \"{word}\"!")
                return
        else:
            guessed_letters.append(guess)
            incorrect_guesses += 1
            print(display_word(word, guessed_letters))
            print(f"Incorrect guesses: {incorrect_guesses}/{max_guesses}")

    print(f"Game over! The word was \"{word}\".")

# Run the hangman game with a maximum of 6 incorrect guesses
hangman(word_list, 6)
```
